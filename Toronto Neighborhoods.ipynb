```python
#Load libraries
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

print('Libraries imported.')
```

```python
# import html-website into a dataframe with pandas
dfs = pd.read_html ('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', header=0)
for df in dfs:
    print (df)[0]
```

```python
# replace "not assigned" to NaN
df.replace("Not assigned", np.nan, inplace = True)
df.head(5)
```

```python
# drop rows with not assigned borough
df.dropna(subset=["Borough"], axis=0, inplace = True)
```

```python
# replace missing values in neighbourhood with the borough
df=df.fillna (value='Borough')
```

```python
# convert 'Neighbourhood' to a string for the following grouping code
df['Neighbourhood']= df['Neighbourhood'].astype('str')
```

```python
# group by 'Postcode' and 'Borough' and join Neighbourhoods into a list separated by ,
df= df.groupby(by=['Postcode','Borough']).agg({'Neighbourhood':lambda x:', '.join(x)})
df
```
